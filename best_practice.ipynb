### 1. Setting up a new experiment
#### 1.1 Preparation
To create a new experiment, a data set has to be added first. Enter the name of the data set in the upper text field and confirm it with the 'Add Dataset' button.
<div align="middle">
<video width="80%" controls>

After adding at least one data set, the options are unlocked. You can navigate through all configuration settings using the tabs at the top (red box). All parameters of the 'General' and 'Model' tabs apply globally to the entire experiment. The configuration for the 'Material', 'Background', 'Cameraviews' and 'Lightning' tabs are selected individually for each data set.

#### 1.4 Material configuration

Here you can select the material for the CAD file and the base plate. The chosen materials are imported from a Blender file (.blend). It is recommended to create a dedicated Blender file as a database for all relevant materials that will be used. Several tutorials for creating a wide variety of surfaces in Blender can be found on the internet (e.g. Youtube search: "blender metal material").<br>
Load the database. The created materials within the selected .blend file can now be assigned to the model and the base plate.<br>
Clicking the 'Apply' button, adds the materials to the objects in Blender.<br>

#### 1.6 Cameraviews configuration

In the 'Cameraviews' tab, the camera positions of the data set are defined. An existing config file can be chosen via the 'Select' field and finally imported clicking the 'Load' button.<br>
There are three available methods to create new camera positions: 'Sperical Coordinates', 'Icosphere' and 'Custom'. Those are explained individually in the following paragraphs. Hit the 'Apply' button to confirm the adjustments. If you want to export the settings, click on 'Save as json'. 

##### 1.6.1 Spherical Coordiantes method

This method is used to calculate the camera positions with spherical coordinates. The camera is rotated around the model incrementally. The parameters for the adjustment of the calculation are detailed below.

##### 1.6.3 Custom method
The 'Custom' method allows you to set the camera positions manually. The x-, y- and z-values are typed into the corresponding fields and the coordinate is appended to the list with the 'Add' button. The arrows can be used to scroll through all entries that have been added so far. The currently selected coordinate can be deleted by using the Delete button. Once you are done, click on 'Apply'.<br>

#### 1.8 Name and location of the experiment
The 'Select' field below the 'Dataset status' can be used to specify the location and the name of the experiment. After confirming, the 'Experiment Directory' turns green. As soon as all 'Dataset status' fields are marked green, the 'Save Experiment' and 'Create Experiment' buttons are enabled.

The 'Save Experiment' button saves the experiment in the specified directory. It can then be rendered on another system (e.g. Elwetritsch).<br>The 'Create Experiment' button also saves the experiment and renders it directly afterwards.<br>
